```python
##câu 11
import numpy as np

# Lọc dữ liệu không thiếu để tính toán hệ số hồi quy
train_data = df.dropna(subset=['Performance', 'Hours_Worked', 'Training_Hours'])

# Xác định các biến độc lập (X) và biến phụ thuộc (y)
X = train_data[['Hours_Worked', 'Training_Hours']]
y = train_data['Performance']

# Tính toán hệ số hồi quy thủ công
X = np.c_[np.ones(X.shape[0]), X]  # Thêm cột 1 vào X để tính hệ số c (intercept)
coefficients = np.linalg.inv(X.T @ X) @ X.T @ y  # Công thức hồi quy tuyến tính
c, a, b = coefficients  # Hệ số c (intercept), a (Hours_Worked), b (Training_Hours)

# Điền giá trị thiếu trong Performance
for index, row in df[df['Performance'].isna()].iterrows():
    df.loc[index, 'Performance'] = c + a * row['Hours_Worked'] + b * row['Training_Hours']

    # Điền giá trị thiếu bằng trung bình của từng Department
df_mean = df.copy()
df_mean['Performance'] = df_mean.groupby('Department')['Performance'].transform(lambda x: x.fillna(x.mean()))

# So sánh giá trị giữa hai phương pháp
comparison = df[['Performance']].copy()
comparison['Performance_Regression'] = df['Performance']
comparison['Performance_Mean'] = df_mean['Performance']

print("So sánh giá trị giữa hồi quy và trung bình:")
print(comparison)

# Cách tiếp cận lai
df_hybrid = df.copy()

for index, row in df_hybrid[df_hybrid['Performance'].isna()].iterrows():
    if not np.isnan(row['Hours_Worked']) and not np.isnan(row['Training_Hours']):
        # Sử dụng hồi quy tuyến tính nếu có đủ dữ liệu
        df_hybrid.loc[index, 'Performance'] = c + a * row['Hours_Worked'] + b * row['Training_Hours']
    else:
        # Sử dụng trung bình nếu thiếu dữ liệu
        department_mean = df_hybrid.groupby('Department')['Performance'].transform('mean')
        df_hybrid.loc[index, 'Performance'] = department_mean[index]

print("Dữ liệu sau khi áp dụng cách tiếp cận lai:")
print(df_hybrid)


```

```python
## CÂU 18
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# Đặt các tham số
batch_size = 64
lr = 0.001
epochs = 20
patience = 5

# Tải dữ liệu MNIST
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
testset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# Chia tập train thành train và validation
train_size = int(0.8 * len(trainset))
val_size = len(trainset) - train_size
train_data, val_data = random_split(trainset, [train_size, val_size])

trainloader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
valloader = DataLoader(val_data, batch_size=batch_size, shuffle=False)
testloader = DataLoader(testset, batch_size=batch_size, shuffle=False)

# Xây dựng kiến trúc ANN
class ANNModel(nn.Module):
    def __init__(self):
        super(ANNModel, self).__init__()
        self.fc1 = nn.Linear(28*28, 128)  # Lớp ẩn 1
        self.fc2 = nn.Linear(128, 64)     # Lớp ẩn 2
        self.fc3 = nn.Linear(64, 10)      # Lớp đầu ra
        self.relu = nn.ReLU()
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        x = x.view(-1, 28*28)  # Flatten 28x28 ảnh thành vector 784 chiều
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Khởi tạo mô hình, loss function và optimizer
model = ANNModel()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=lr)

# Early stopping
best_loss = float('inf')
patience_counter = 0
train_losses = []
val_losses = []

# Huấn luyện mô hình
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for images, labels in trainloader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    avg_train_loss = running_loss / len(trainloader)
    train_losses.append(avg_train_loss)

    # Validation
    model.eval()
    running_val_loss = 0.0
    with torch.no_grad():
        for images, labels in valloader:
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_val_loss += loss.item()

    avg_val_loss = running_val_loss / len(valloader)
    val_losses.append(avg_val_loss)

    # Kiểm tra early stopping
    if avg_val_loss < best_loss:
        best_loss = avg_val_loss
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch + 1}")
            break

    print(f"Epoch [{epoch + 1}/{epochs}], Train Loss: {avg_train_loss:.4f}, Validation Loss: {avg_val_loss:.4f}")

# Vẽ biểu đồ loss
plt.plot(range(len(train_losses)), train_losses, label='Training Loss')
plt.plot(range(len(val_losses)), val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()

# Đánh giá trên tập test
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for images, labels in testloader:
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f'Accuracy on test set: {accuracy:.2f}%')

```

```python
## CÂU 19
from sklearn.metrics import accuracy_score, confusion_matrix
import torch
from torch.utils.data import DataLoader

# Đánh giá mô hình trên tập test
model.eval()
testloader = DataLoader(testset, batch_size=64, shuffle=False)

# Lưu trữ các nhãn thực tế và dự đoán của mô hình
all_labels = []
all_preds = []

# Bắt đầu đánh giá mô hình
with torch.no_grad():
    for images, labels in testloader:
        outputs = model(images)
        _, preds = torch.max(outputs, 1)  # Lấy nhãn dự đoán có xác suất cao nhất
        all_labels.extend(labels.numpy())
        all_preds.extend(preds.numpy())

# Tính accuracy và kiểm tra giá trị accuracy
accuracy = accuracy_score(all_labels, all_preds)
print(f"Accuracy on test set: {accuracy * 100:.2f}%")  # In ra giá trị accuracy

# Nếu accuracy dưới 95%, phân tích confusion matrix
if accuracy < 0.95:
    print("Accuracy dưới 95%, tiếp tục phân tích confusion matrix...")  # Thông báo khi accuracy dưới 95%
    cm = confusion_matrix(all_labels, all_preds)

    # Phân tích các chữ số bị nhầm lẫn nhiều nhất
    misclassified = {}
    for i in range(10):
        misclassified[i] = cm[i].sum() - cm[i, i]  # Tổng số lỗi cho mỗi lớp

    # In ra các chữ số bị nhầm lẫn nhiều nhất
    sorted_misclassified = sorted(misclassified.items(), key=lambda x: x[1], reverse=True)
    print("\nChữ số bị nhầm lẫn nhiều nhất:")
    for digit, count in sorted_misclassified[:5]:  # Hiển thị 5 chữ số bị nhầm lẫn nhiều nhất
        print(f"Chữ số {digit}: {count} lần nhầm lẫn")

    # Đề xuất cải tiến mô hình
    print("\nĐề xuất cải tiến mô hình:")
    print("- Mô hình có thể bị nhầm lẫn giữa các chữ số tương tự nhau như 3 và 8, 4 và 9.")
    print("- Cải tiến có thể bao gồm việc sử dụng các mạng CNN thay vì ANN, giúp mô hình học tốt hơn các đặc trưng hình học của chữ số.")
    print("- Ngoài ra, tăng cường dữ liệu (data augmentation) như xoay hoặc biến đổi hình ảnh có thể giúp mô hình phân biệt tốt hơn giữa các chữ số tương tự.")
else:
    print("Accuracy đã đạt 95% hoặc cao hơn, không cần phân tích confusion matrix.")


```
